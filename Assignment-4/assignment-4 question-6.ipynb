```python
#empty lists
Song_name =[]
Artist_name =[]
Lastweek_rank =[]
Peak_rank =[]
Weekson_board =[]
```

```python
#scraping song name
name_tags=driver.find_elements(By.XPATH,'//li[@class="o-chart-results-list__item // lrv-u-flex-grow-1 lrv-u-flex lrv-u-flex-direction-column lrv-u-justify-content-center lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light lrv-u-padding-l-050 lrv-u-padding-l-1@mobile-max"]')
for i in name_tags[0:100]:
    name=i.text
    Song_name.append(name)
```

```python
#scarping artist name
name_tags=driver.find_elements(By.XPATH,'//span[@class="c-label  a-no-trucate a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block a-truncate-ellipsis-2line u-max-width-330 u-max-width-230@tablet-only"]')
for i in name_tags[0:100]:
    name=i.text
    Artist_name.append(name)
```
